```python
import numpy as np
import pandas as pd
import sys
import time
import pickle 

start = time.process_time()


def add_time_variables(): 
    
    # the function has 2 purposes:
    # 1 - to create a numeric time variable (e.g. year_month instead of a new one)
    # 2 - to create binary columns for items that existed in the last 3 and 9 months
    
    # creating a time variable
    wds.year_month.replace('/', '', regex=True, inplace=True)
    wds.sort_values(by=['year_month'], inplace=True)
    
    # getting the last 3 and 9 months as values in lists
    last_3_months=sorted(wds.year_month.unique().tolist(),reverse=True)
    last_9_months=sorted(wds.year_month.unique().tolist(),reverse=True)
    del last_3_months[3:]
    del last_9_months[9:]
    
    # creating the two new columns
    wds.loc[wds['year_month'].isin(last_3_months), "last_3_months"] = 1
    wds.loc[wds['year_month'].isin(last_9_months), "last_9_months"] = 1
    wds.fillna(0, inplace=True)
    
    # ordering the df -> most recent values on the top
    wds.sort_values(by=['year_month'], ascending=False, inplace=True)
    wds.reset_index(inplace=True, drop=True)
    

wds=pd.read_csv(r"C:\Users\n0l\Desktop\Pandas\csvForTest\inputCSV.csv", sep=';', engine='python')
#wds=pd.read_csv(r"C:\Users\n0l\Desktop\Pandas\csvForTest\1_monthCSV.csv", sep=';', engine='python')

#shuffle ds
wds = wds.sample(frac=1).reset_index(drop=True)
#print(id(wds))

add_time_variables()

print(time.process_time() - start)
wds
```
